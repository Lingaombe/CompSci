```python
# analysis of daily sales in a retail

import pandas as pd

daily_sales = [1500, 1820, 2500, 1000, 2500, 1500, 1680, 1820, 2100, 2200, 1800, 1700, 1600, 2100, 8200]

sales_data = pd.Series(daily_sales)

mean_sales = sales_data.mean()
median_sales = sales_data.median()
mode_sales = sales_data.mode()
var_sales = sales_data.var()
std_sales = sales_data.std()
min_sales = sales_data.min()
max_sales = sales_data.max()
range_sales = max_sales - min_sales

print(f"Mean Sales: {mean_sales:.2f} Rs")
print(f"Median Sales: {median_sales:.2f} Rs")
print(f"Mode Sales: {mode_sales.tolist()} Rs")
print(f"Variance: {var_sales:.2f} Rs²")
print(f"Standard Deviation: {std_sales:.2f} Rs")
print(f"Minimum Sales: {min_sales} Rs")
print(f"Maximum Sales: {max_sales} Rs")
print(f"Range of Sales: {range_sales} Rs")
```
